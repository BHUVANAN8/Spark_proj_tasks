##**Objective:** Create the Decision Tree classifier and visualize it graphically.

## **Importing libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
print("setup complete")
```

## **Prepare the data**

```python
#Loading the data
df=datasets.load_iris()
#forming the feauture variable and target variable
X=pd.DataFrame(df.data, columns=df.feature_names)
Y=df.target
```

```python
#print the X and Y values
print(X)
print(Y)
```

```python
#first 5 rows
X.head()
```

## **BUILD A MODEL USING DECISION TREE CLASSIFIER**

```python
# Fit the classifier with default hyper-parameters
clf = DecisionTreeClassifier(random_state=120)
model = clf.fit(X, Y)
print('model Built sucessfully')
```

## **VISUALIZE THE TREE**

```python
#plot the tree
plt.figure(figsize=(25,20))
Decision_tree = tree.plot_tree(clf,feature_names=df.feature_names,class_names=df.target_names,filled=True)
```

## **Here am using plot_tree method to plot the figure of the tree and plot_tree returns the annotations so am here declared the variable Decision_tree to not show them and notice that filled = true in the plot_tree. When this parameter is set to True the method uses color to indicate the majority of the class.**
